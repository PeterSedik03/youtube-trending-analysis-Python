# 1. Create a single dataframe with the concatenation of all input csv files, adding a column called country

```python
import pandas as pd
import pathlib as pl
folder = pl.Path(r"C:\Users\peter_6\Downloads\trendingYT\trendingYT")
df = pd.concat(( pd.read_csv(f, encoding="utf-8", encoding_errors="replace")
                .assign(country=f.stem[:2])
        for f in folder.glob("*.csv")),ignore_index=True)
df.head()
```

# 2. Extract all videos that have no tag.

```python
no_tags = df[df['tags'].isna() | (df['tags'] == '[none]')]
no_tags
```

# 3. For each channel, determine the total number of views

```python
channel_n_views = (df.drop_duplicates(["video_id", "country"])
      .groupby("channel_title", as_index=False)["views"].sum()
      .sort_values("views", ascending=False))
channel_n_views
```

# 4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded, and remove those rows from the original dataframe.

```python
excluded = df[((df['comments_disabled']) & (df['ratings_disabled'])) |
    (df['video_error_or_removed'])]
```

```python
df = df.drop(excluded.index)
df.head()
```

```python
df.shape
```

# 5. Add a like_ratio column storing the ratio between the number of likes and dislikes

```python
df.head()
```

```python
import numpy as np
df["like_ratio"] = np.where(df["dislikes"] > 0, df["likes"] / df["dislikes"], np.nan)
cols = list(df.columns)
cols.insert(cols.index('dislikes') + 1, cols.pop(cols.index('like_ratio')))
df = df[cols]
```

```python
df.head()
```

# 6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
df['publish_time_dt'] = pd.to_datetime(df['publish_time'])
df['publish_time_interval'] = df['publish_time_dt'].dt.floor('10min')
df['publish_time_interval_label'] = (
    df['publish_time_interval'].dt.strftime('%H:%M') + '-' +
    (df['publish_time_interval'] + pd.Timedelta(minutes=10)).dt.strftime('%H:%M'))
```

```python
df[['publish_time', 'publish_time_dt', 'publish_time_interval', 'publish_time_interval_label']]
```

# 7. For each interval, determine the number of videos, average number of likes and of dislikes.

```python
interval_stats = (
    df.drop_duplicates(["video_id", "country"]).groupby("publish_time_interval")
      .agg(num_videos = ("video_id","count"), avg_likes = ("likes","mean"),
           avg_dislikes = ("dislikes","mean")).reset_index())
interval_stats
```

# 8. For each tag, determine the number of videos
Notice that tags contains a string with several tags.

```python
df_vid = df.drop_duplicates(["video_id", "country"])
tags_exploded = (df_vid[df_vid["tags"] != "[none]"]["tags"]
      .str.replace('"', '', regex=False).str.split("|").explode().dropna().str.strip())
tag_n_videos = (tags_exploded.value_counts().reset_index()
                .rename(columns={"tags": "tag", "count": "num_videos"}))
if "index" in tag_n_videos.columns:
    tag_n_videos = tag_n_videos.rename(columns={"index": "tag", "tags": "num_videos"})
tag_n_videos
```

# 9. Find the tags with the largest number of videos

```python
popular_tags = (tag_n_videos.sort_values("num_videos", ascending=False).head(10)
      .reset_index(drop=True))
```

```python
popular_tags
```

# 10. For each (tag, country) pair, compute average ratio likes/dislikes

```python
df.head()
```

```python
df_vid = df.drop_duplicates(["video_id", "country"]).copy()
df_exploded = df_vid[df_vid["tags"] != "[none]"].copy()
df_exploded["tag"] = (df_exploded["tags"].str.replace('"', '', regex=False).str.split("|"))
df_exploded = df_exploded.explode("tag")
df_exploded["tag"] = df_exploded["tag"].astype(str).str.strip()
df_exploded = df_exploded[df_exploded["tag"] != ""]
```

```python
avg_tag_country = (df_exploded.groupby(["tag", "country"])["like_ratio"].mean().reset_index())
avg_tag_country
```
