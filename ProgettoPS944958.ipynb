# 1. Create a single dataframe with the concatenation of all input csv files, adding a column called country

```python
import pandas as pd
import pathlib as pl
cartella = pl.Path(r"C:\Users\peter_6\Downloads\trendingYT\trendingYT")
df = pd.concat((pd.read_csv(f, encoding="latin1").assign(country=f.stem[:2]) 
     for f in cartella.glob("*.csv")), ignore_index=True)
df.head()
```

# 2. Extract all videos that have no tag.

```python
no_tags = df[df['tags'] == '[none]']
no_tags
```

# 3. For each channel, determine the total number of views

```python
channel_n_views = df.groupby("channel_title", as_index=False)["views"].sum().sort_values("views", ascending=False)
```

```python
channel_n_views
```

# 4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded, and remove those rows from the original dataframe.

```python
excluded_df = df[(df['comments_disabled']) & (df['ratings_disabled']) | (df['video_error_or_removed'])]
```

```python
df = df.drop(excluded_df.index)
df.head()
```

```python
df.shape
```

# 5. Add a like_ratio column storing the ratio between the number of likes and dislikes

```python
df.head()
```

```python
import numpy as np
df["like_ratio"] = np.where(df["dislikes"] > 0, df["likes"] / df["dislikes"], np.nan)
cols = list(df.columns)
cols.insert(cols.index('dislikes') + 1, cols.pop(cols.index('like_ratio')))
df = df[cols]
```

```python
df.head()
```

# 6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])
df['publish_time_interval'] = df['publish_time'].dt.floor('10min')
```

```python
df[['publish_time', 'publish_time_interval']]
```

# 7. For each interval, determine the number of videos, average number of likes and of dislikes.

```python
interval_stats = (df.drop_duplicates(["video_id", "country"]) 
      .groupby("publish_time_interval").agg(num_videos=("video_id", "nunique"),
          avg_likes=("likes", "mean"),avg_dislikes=("dislikes", "mean"),).reset_index())
```

```python
interval_stats
```

# 8. For each tag, determine the number of videos
Notice that tags contains a string with several tags.

```python
df_vid = df.drop_duplicates(["video_id", "country"])
tags_exploded = (df_vid[df_vid["tags"] != "[none]"]["tags"].str.replace('"', '', regex=False)
      .str.split("|").explode().dropna().str.strip())
tag_n_videos = tags_exploded.value_counts().reset_index()
tag_n_videos.columns = ["tag", "num_videos"]
```

```python
tag_n_videos
```

# 10. For each (tag, country) pair, compute average ratio likes/dislikes

```python
df_vid = df.drop_duplicates(["video_id", "country"]).copy()
df_vid["like_ratio"] = np.where(df_vid["dislikes"] > 0, df_vid["likes"]/df_vid["dislikes"], np.nan)
df_exploded = df_vid[df_vid["tags"] != "[none]"].copy()
df_exploded["tag"] = (df_exploded["tags"].str.replace('"','', regex=False).str.split("|"))
df_exploded = df_exploded.explode("tag")
df_exploded["tag"] = df_exploded["tag"].astype(str).str.strip()
df_exploded = df_exploded[df_exploded["tag"] != ""]
```

```python
avg_tag_country = (df_exploded.groupby(["tag","country"])["like_ratio"].mean().reset_index())
```

```python
avg_tag_country
```

# 12. Divide trending_date into three columns: year, month, day


```python
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
df['year'] = df['trending_date'].dt.year
df['month'] = df['trending_date'].dt.month
df['day'] = df['trending_date'].dt.day
```

# 13. For each (month, country) pair, the video with the largest number of views

```python
monthly_max_n_views = df.loc[df.groupby(["month", "country"])["views"].idxmax()]
```

# 14. Read all json files with the video categories

```python
import json
cartella = pl.Path(r"C:\Users\peter_6\Downloads\trendingYT\trendingYT")
dfjs = pd.concat((pd.json_normalize(json.loads(f.read_text(encoding="utf-8"))["items"])
        .assign(country=f.stem.split("_")[0])
                  for f in cartella.glob("*_category_id*.json")), ignore_index=True)
dfjs['id'] = dfjs['id'].astype(int)
dfjs.head()
```

```python
dfjs.shape
```

# 15. For each country, determine how many videos have a category that is not assignable.

```python
df_merged = pd.merge(df, dfjs,left_on=["category_id", "country"],
    right_on=["id", "country"],how="left")
not_assignable = df_merged[df_merged["snippet.assignable"] == False]
risultato_15 = not_assignable.groupby("country").size()
risultato_15_unique = not_assignable.groupby("country")["video_id"].nunique()
```
